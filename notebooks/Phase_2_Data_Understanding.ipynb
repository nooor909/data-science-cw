## 5. Applications Over Time + Approval Rate Over Time

```python
df['project_submitted_datetime'] = pd.to_datetime(df['project_submitted_datetime'])
df['submission_month'] = df['project_submitted_datetime'].dt.to_period('M')

monthly_counts = df['submission_month'].value_counts().sort_index()
monthly_counts.index = monthly_counts.index.to_timestamp()
approval_over_time = df.groupby('submission_month')['project_is_approved'].mean()
approval_over_time.index = approval_over_time.index.to_timestamp()

# Applications Over Time
plt.figure(figsize=(12, 5))
monthly_counts.plot()

plt.title("Applications Submitted Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Applications")
plt.tight_layout()
plt.show()

# Approval Rate Over Time
plt.figure(figsize=(12, 5))
approval_over_time.plot()

plt.title("Approval Rate Over Time")
plt.xlabel("Month")
plt.ylabel("Approval Rate")
plt.tight_layout()
plt.show()
```

#### Insights on Applications Submitted Over Time

- There is a sharp increase from April until August.
- August had the most Applications overall with roughly 34,000 applications.
- This may suggest a seasonal or cyclical pattern. That spring and summer have the majority of applications.
- This may be due to funding deadlines, school cycles, or academic calander.

#### Insights on Approval Rate Over Time

- Approval starts at a high rate in April.
- It starts decreasing gradually until it reaches a low around September.
- Rate stays at an average 85% and 86% approval roughly for around 4 months starting from November up until March.

#### Key Insights
- When applications increase, then approval rate drops.
- When applications decrease, then approval rate slightly rises.
- There is a negative relationship and correlation between the number of applications alongside approval rates.

## Missing Values Analysis

```python
# Check for missing values.

missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_report = pd.DataFrame({
    'Missing Count': missing, 
    'Missing %': missing_pct})

missing_report = missing_report[missing_report['Missing Count'] > 0]
missing_report = missing_report.sort_values(by="Missing %", ascending=False)

print("Missing Values Report: ")
print(missing_report)
```

```python
df[df['teacher_prefix'].isnull()]
```

```python
# Displaying the missing values in every column
missing = df.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]

plt.figure(figsize=(10,5))
missing.head(15).plot(kind='bar')
plt.title("Top Columns with Missing Values")
plt.ylabel("Number of Missing Values")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

#### Insights on Missing Data:

## Detecting Outliers

```python
# Detect outliers using box plots or statistical methods.

# Teacher Experience Outliers

Q1 = df['teacher_number_of_previously_posted_projects'].quantile(0.25)
Q3 = df['teacher_number_of_previously_posted_projects'].quantile(0.99)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[
    (df['teacher_number_of_previously_posted_projects'] < lower_bound) |
    (df['teacher_number_of_previously_posted_projects'] > upper_bound)
]

print("Lower bound: ", lower_bound)
print("Upper bound: ", upper_bound)
print("Number of outliers: ", len(outliers))
outliers.head()
```

#### Insight on Outlier:

- The feature teacher_number_of_previously_posted_projects contains significant outliers.
- A small number of teachers have submitted a very large number of projects compared to the majority.
- These outliers may influence model performance and may require treatment.
