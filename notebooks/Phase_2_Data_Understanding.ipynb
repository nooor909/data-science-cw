### Task 3: Explore Data

Dig deeper into the data using querying, visualisation, and statistical analysis to find initial insights, patterns, and interesting subsets. Key activities include:

- **Univariate Analysis:** Examine the distribution of individual features using histograms, box plots, and bar charts.
- **Bivariate / Multivariate Analysis:** Explore relationships between features using scatter plots, correlation matrices, and grouped summaries.

**Instructions:** Create visualisations and run queries to understand the distributions, relationships, and patterns in your data.

```python
# Perform univariate analysis — explore distributions of individual features.

# Numerical Distribution
df.select_dtypes(include='number').hist(figsize=(14, 10), bins=30)
plt.suptitle("Distribution of Numerical Features")
plt.show()

# Target Variable
sns.countplot(x='project_is_approved', data=df)
plt.title("Distribution of Project Approval")
plt.show()
```

- The target variable which is project_is_approved' is imbalanced, with significantly more approved applications than rejected ones.
- The feature teacher_number_of_previously_posted_projects is highly skewed, with most teachers having submitted very few projects and a small number having many projects submitted.
- This indicates the presence of outliers and varying levels of teacher experience.
- The imbalance in the target variable suggests that model performance metrics must be chosen carefully in later stages.

```python
# Perform bivariate/multivariate analysis — explore relationships between features.

# Approval vs Teacher Experience
sns.boxplot(x='project_is_approved', 
            y='teacher_number_of_previously_posted_projects', 
            data=df)
plt.title("Approval vs Teacher Experience")
plt.show()

# Correlation Matrix
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

- Teachers with more previously posted projects tend to show a slightly higher likelihood of approval, suggesting experience may influence outcomes.
- Futher analysis of categorical features may provide deeper insights.
- No strong linear correlations were observed between numerical variables, indicating that approval decisions are likely influenced by multiple factors rather than a single dominant variable.

### Task 4: Verify Data Quality

Evaluate the completeness and correctness of the data. Check for problems that could affect the analysis:

- **Missing Values:** Identify columns with missing data, the volume, and patterns of missingness.
- **Duplicates:** Check for and report duplicate records.
- **Outliers:** Detect anomalous values using statistical methods or visualisations (box plots, z-scores).
- **Inconsistencies:** Look for data type mismatches, inconsistent formatting, or invalid entries.

**Instructions:** Systematically check for data quality issues and document your findings.

```python
# Check for missing values.

missing = df.isnull().sum()

missing_pct = (missing / len(df)) * 100

missing_report = pd.DataFrame({
    'Missing Count': missing, 
    'Missing %': missing_pct})

missing_report = missing_report[missing_report['Missing Count'] > 0]
missing_report = missing_report.sort_values(by="Missing %", ascending=False)

print("Missing Values Report: ")
print(missing_report)
```

Missing Data Notes:
- Several columns contain missing values, though most are relatively small in proportion.
- Missing data may affect model performance if not handled properly.
- These values will need to be addressed in the Data Preparation phase.

```python
# Check for duplicate records.

duplicates = df.duplicated().sum()

print(f"Number of duplicate rows: {duplicates}")

if duplicates > 0:
    print("\nSample of duplicate rows:")
    print(df[df.duplicated()].head())
```

Duplicate Data Notes:
- No significant duplicate records were found in the dataset.


```python
# Detect outliers using box plots or statistical methods.

numerical_cols = df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    plt.figure()
    plt.boxplot(df[col].dropna())
    plt.title(f"Boxplot of {col}")
    plt.show()
```

Outlier Notes:
- The feature teacher_number_of_previously_posted_projects contains significant outliers.
- A small number of teachers have submitted a very large number of projects compared to the majority.
- These outliers may influence model performance and may require treatment.

```python
# Check unique values in the different columns

categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df[col].value_counts().head(5))
```

Data Consistency Notes:
- Categorical variables appear generally consistent, with no obvious formatting issues in the sampled values.
- Some categorical features contain many unique value, which may require dimensionality reduction or grouping.
